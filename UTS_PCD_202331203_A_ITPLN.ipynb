```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

<a id="import"></a>
<div style="margin-top:30px; padding:10px; background-color:#FF9F00; border-left: 10px solid #CB0404; border-radius:5px">
    <h2 style="color:white; margin:0">2. Splitting Kanal Warna</h2>
</div>

<!-- Heading 2 -->
<div style="margin-top:30px; padding:10px; background-color:#F4631E; border-left: 10px solid #309898; border-radius:5px">
    <h3 style="color:white; margin:0">2.1. Memasukkan Citra</h3>
</div>

```python
citra_bgr = cv2.imread('nama.jpg')
citra_rgb = cv2.cvtColor(citra_bgr, cv2.COLOR_BGR2RGB)
```

<!-- Heading 2 -->
<div style="margin-top:30px; padding:10px; background-color:#F4631E; border-left: 10px solid #309898; border-radius:5px">
    <h3 style="color:white; margin:0">2.2. Splitting Citra</h3>
</div>

```python
r = citra_rgb[:,:,0]
g = citra_rgb[:,:,1]
b = citra_rgb[:,:,2]
```

<!-- Heading 2 -->
<div style="margin-top:30px; padding:10px; background-color:#F4631E; border-left: 10px solid #309898; border-radius:5px">
    <h3 style="color:white; margin:0">2.3. Konversi Grayscale (Manual)</h3>
</div>

```python
citra_gray = (0.2989 * r + 0.587 * g + 0.1141 * b).astype(np.uint8)
```

<!-- Heading 2 -->
<div style="margin-top:30px; padding:10px; background-color:#F4631E; border-left: 10px solid #309898; border-radius:5px">
    <h3 style="color:white; margin:0">2.4. Inisiasi Histogram Setiap Kanal</h3>
</div>

```python
hist_r = cv2.calcHist([r], [0], None, [256], [0, 256])
hist_g = cv2.calcHist([g], [0], None, [256], [0, 256])
hist_b = cv2.calcHist([b], [0], None, [256], [0, 256])
hist_gray = cv2.calcHist([citra_gray], [0], None, [256], [0, 256])
```

<!-- Heading 2 -->
<div style="margin-top:30px; padding:10px; background-color:#F4631E; border-left: 10px solid #309898; border-radius:5px">
    <h3 style="color:white; margin:0">2.5. Tampilan Citra dan Perbandingan Statistik Kanal Warna</h3>
</div>

<!-- Heading 2 -->
<div style="margin-top:30px; padding:10px; background-color:#F4631E; border-left: 10px solid #309898; border-radius:5px">
    <h3 style="color:white; margin:0">3.1. Memasukan Citra</h3>
</div>

```python
fig, axs = plt.subplots(5, 2, figsize=(12, 15))

# 1. RGB + Histogram RGB
axs[0, 0].imshow(citra_rgb)
axs[0, 0].set_title("Citra RGB")
axs[0, 0].axis('off')
axs[0, 1].plot(hist_r, color='red', label='Red')
axs[0, 1].plot(hist_g, color='green', label='Green')
axs[0, 1].plot(hist_b, color='blue', label='Blue')
axs[0, 1].set_title("Histogram RGB")
axs[0, 1].legend()

# 2. Red Channel + Histogram
axs[1, 0].imshow(r, cmap='gray')  
axs[1, 0].set_title("Merah")
axs[1, 0].axis('off')
axs[1, 1].plot(hist_g, color='green', label='Green')
axs[1, 1].plot(hist_b, color='blue', label='Blue')
axs[1, 1].set_title("Histogram Green & Blue")
axs[1, 1].legend()

# 3. Green Channel + Histogram
axs[2, 0].imshow(g, cmap='gray')
axs[2, 0].set_title("Hijau")
axs[2, 0].axis('off')
axs[2, 1].plot(hist_b, color='blue', label='Blue')
axs[2, 1].plot(hist_r, color='red', label='Red')
axs[2, 1].set_title("Histogram Red & Blue")
axs[2, 1].legend()

# 4. Blue Channel + Histogram
axs[3, 0].imshow(b, cmap='gray') 
axs[3, 0].set_title("Biru")
axs[3, 0].axis('off')
axs[3, 1].plot(hist_r, color='red', label='Red')
axs[3, 1].plot(hist_g, color='green', label='Green')
axs[3, 1].set_title("Histogram Red & Green")
axs[3, 1].legend()

# 5. Grayscale + Histogram
axs[4, 0].imshow(citra_gray, cmap='gray')
axs[4, 0].set_title("Citra Grayscale")
axs[4, 0].axis('off')
axs[4, 1].plot(hist_gray, color='black')
axs[4, 1].set_title("Histogram Grayscale")

plt.tight_layout()
plt.show()
```

```python
citra_bgr = cv2.imread('nama.png')
citra_binary = cv2.cvtColor(citra_bgr, cv2.COLOR_BGR2RGB)
```

<!-- Heading 2 -->
<div style="margin-top:30px; padding:10px; background-color:#F4631E; border-left: 10px solid #309898; border-radius:5px">
    <h3 style="color:white; margin:0">3.2. Splitting Citra</h3>
</div>

```python
r = citra_binary[:,:,0]
g = citra_binary[:,:,1]
b = citra_binary[:,:,2]
```

<!-- Heading 2 -->
<div style="margin-top:30px; padding:10px; background-color:#F4631E; border-left: 10px solid #309898; border-radius:5px">
    <h3 style="color:white; margin:0">3.3. Conversi Grayscale (Manual)</h3>
</div>

```python
threshold_r = 0
threshold_g = 0
threshold_b = 0

binary1 = np.where((b > r) & (b > g) & (b > threshold_b), 255, 0).astype(np.uint8)

binary2 = np.where(
    (((b > g) & (b > r) & (b > threshold_b)) |
     ((r > g) & (r >= b) & (r > threshold_r))),
    255, 0).astype(np.uint8)

binary3 = np.where(
    (((r > g) & (r > b) & (r > threshold_r)) |
     ((g > r) & (g > b) & (g > threshold_g)) |
     ((b > r) & (b > g) & (b > threshold_b))),
    255, 0).astype(np.uint8)

binary0 = np.zeros_like(r)
```

<!-- Heading 2 -->
<div style="margin-top:30px; padding:10px; background-color:#F4631E; border-left: 10px solid #309898; border-radius:5px">
    <h3 style="color:white; margin:0">3.4. Inisiasi Histogram Setiap Kanal</h3>
</div>

```python
hist_bin0 = cv2.calcHist([binary0], [0], None, [256], [0, 256])
hist_bin1 = cv2.calcHist([binary1], [0], None, [256], [0, 256])
hist_bin2 = cv2.calcHist([binary2], [0], None, [256], [0, 256])
hist_bin3 = cv2.calcHist([binary3], [0], None, [256], [0, 256])
```

<!-- Heading 2 -->
<div style="margin-top:30px; padding:10px; background-color:#F4631E; border-left: 10px solid #309898; border-radius:5px">
    <h3 style="color:white; margin:0">3.5. Tampilan Citra dan Perbandingan Statistik Kanal Warna</h3>
</div>

```python
fig2, axs2 = plt.subplots(2, 2, figsize=(12, 10))

axs2[0, 0].imshow(binary0, cmap='gray')
axs2[0, 0].set_title("NONE")
axs2[0, 0].axis('off')

axs2[0, 1].imshow(binary1, cmap='gray')
axs2[0, 1].set_title("BLUE")
axs2[0, 1].axis('off')

axs2[1, 0].imshow(binary2, cmap='gray')
axs2[1, 0].set_title("RED-BLUE")
axs2[1, 0].axis('off')

axs2[1, 1].imshow(binary3, cmap='gray')
axs2[1, 1].set_title("RED-GREEN-BLUE")
axs2[1, 1].axis('off')

plt.tight_layout()
plt.show()
```

```python
img = cv2.imread('buram.png')
#202331203_Michael Christia Putra
```

```python
bgr = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
#202331203_Michael Christia Putra
rgb = cv2.cvtColor(bgr, cv2.COLOR_RGB2GRAY)
```

```python
[baris, kolom] = img.shape[:2]
#202331203_Michael Christia Putra
```

```python
beta = 45
citra_cerah = np.zeros((baris, kolom, 3))

for x in range(baris):
    for y in range(kolom):
        gyx = rgb[x,y] + beta
        citra_cerah[x,y] = gyx

citra_cerah = citra_cerah.astype(np.uint8)

fig, axs = plt.subplots(2,2, figsize=(15,5))
axs[0,0].imshow(rgb, cmap='gray')
axs[0,1].hist(rgb.ravel(),256,[0,256])
axs[1,0].imshow(citra_cerah)
axs[1,1].hist(citra_cerah.ravel(),256,[0,256])

plt.show()
#202331203_Michael Christia Putra
```

```python
alpha = 1.2
citra_kontras = np.zeros((baris, kolom, 3))

for x in range(baris):
    for y in range(kolom):
        gyx = rgb[x,y] * alpha
        citra_kontras[x,y] = gyx

citra_kontras = citra_kontras.astype(np.uint8)

fig, axs = plt.subplots(2,2, figsize=(15,5))
axs[0,0].imshow(rgb, cmap='gray')
axs[0,1].hist(rgb.ravel(),256,[0,256])
axs[1,0].imshow(citra_kontras)
axs[1,1].hist(citra_kontras.ravel(),256,[0,256])

plt.show()
#202331203_Michael Christia Putra
```

```python
beta = 10
alpha = 1.2
citra_hasil = np.zeros((baris, kolom, 3))

for x in range(baris):
    for y in range(kolom):
        gyx = alpha * rgb[x,y] + beta
        citra_hasil[x,y] = gyx

fig, axs = plt.subplots(2,2, figsize=(15,5))
citra_hasil = citra_hasil.astype(np.uint8)
axs[0,0].imshow(rgb, cmap='gray')
axs[0,1].hist(rgb.ravel(),256,[0,256])
axs[1,0].imshow(citra_hasil)
axs[1,1].hist(citra_hasil.ravel(),256,[0,256])

plt.show()
#202331203_Michael Christia Putra
```
